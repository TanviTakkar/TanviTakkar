```python
import pandas as pd
data=pd.read_csv("/content/rainfall_in_india_1901-2015.csv")
data.head()
```

```python
data.describe()
```

```python
data.loc[1:10]
```

```python
data0=data.groupby('SUBDIVISION').sum()
data0
```

```python
import matplotlib.pyplot as plt
data2=data0['ANNUAL'].plot(figsize=(21,12),color='red',linewidth=3)
plt.xlabel('SUBDIVISION',color='blue',fontsize=20)
plt.ylabel('ANNUAL',color='blue',fontsize=20)
plt.title('AVG. RAINFALL',color='black',fontsize=25)
data2
plt.grid()
plt.legend()
```

```python
plt.figure(dpi=300)
data3=data.groupby('SUBDIVISION').sum()['ANNUAL'].plot.barh(stacked=True,color='green',fontsize=3)
plt.ylabel("SUBDIVISION",color='red')
plt.xlabel("ANNUAL",color='red')
plt.grid(alpha=0.2)
plt.title('ANNUAL RAINFALL',color='red')
data3
```

```python
plt.figure(dpi=250)
data4=data[['YEAR','JAN','FEB','MAR','APR','MAY','JUN','JUL','AUG','SEP','OCT','NOV','DEC']].groupby('YEAR').sum()
plot1=data4.plot(figsize=(12,12),linewidth=2)
plt.xlabel('YEAR',color='green',fontsize=20)
plt.title('AVG. MONTHLY RAINFALL',color='green',fontsize=29)
plt.grid()
plot1
```

```python
plt.figure(dpi=250)
plot2=data[['YEAR','JAN','FEB','MAR','APR','MAY','JUN','JUL','AUG','SEP','OCT','NOV','DEC']].groupby('YEAR').sum().plot.barh(figsize=(12,24),stacked=True)
plt.ylabel('YEAR',color='red',fontsize=26)
plt.title('MONTHLY AVG. RAINFALL',color='red',fontsize=29)
plt.grid()
plot2
```

```python
data5=data[['JAN','FEB','MAR','APR','MAY','JUN','JUL','AUG','SEP','OCT','NOV','DEC','SUBDIVISION']].groupby('SUBDIVISION').sum()
data5.plot(figsize=(20,12),linewidth=2)
plt.xlabel('SUBDIVISION',color='blue',fontsize=23)
plt.grid()
```

```python
plot3=data[['JAN','FEB','MAR','APR','MAY','JUN','JUL','AUG','SEP','OCT','NOV','DEC','SUBDIVISION']].groupby('SUBDIVISION').sum().plot.barh(figsize=(12,12),stacked=True)
plt.ylabel('SUBDIVISION',color='blue',fontsize=23)
plt.grid()
plot3
```

```python
plt.figure(dpi=250)
data6=data[['SUBDIVISION','Jan-Feb','Mar-May','Jun-Sep','Oct-Dec']].groupby('SUBDIVISION').sum()
data6.plot.barh(figsize=(14,14),stacked=True)
plt.ylabel('SUBDIVISION',color='magenta',fontsize=23)
plt.title('MONTHLY AVG. RAINFALL',color='magenta',fontsize=29)
plt.grid()
```

```python
plt.figure(dpi=400)
plt.figure(figsize=(110,12),linewidth=45)
data6=data[['SUBDIVISION','Jan-Feb','Mar-May','Jun-Sep','Oct-Dec']].groupby('SUBDIVISION').sum()
plt.plot(data6)
plt.xlabel('SUBDIVISION',color='magenta',fontsize=23)
plt.title('MONTHLY AVG. RAINFALL',color='magenta',fontsize=29)
```

```python
plt.figure(dpi=250)
data7=data[['YEAR','Jan-Feb','Mar-May','Jun-Sep','Oct-Dec']].groupby('YEAR').sum()
data7.plot(figsize=(12,12),linewidth=2,marker='o')
plt.legend(loc="upper right",bbox_to_anchor=(1.13,1),facecolor='yellow')
plt.xlabel('YEAR',color='black',fontsize=23)
plt.grid()
```

```python
plot4=data[['YEAR','Jan-Feb','Mar-May','Jun-Sep','Oct-Dec']].groupby('YEAR').sum().plot.barh(stacked=True,figsize=(12,24))
plt.ylabel('YEAR',color='black',fontsize=23)
plt.grid()
plot4
```

```python
Punjab=data.loc[(data['SUBDIVISION']=='PUNJAB')]
Punjab=Punjab.head(20)
Punjab
```

```python
data8=Punjab[['JAN','FEB','MAR','APR','MAY','JUN','JUL','AUG','SEP','OCT','NOV','DEC','YEAR']].groupby('YEAR').sum()
data8.plot.bar(figsize=(12,12),stacked=True)
plt.title('MONTHLY RAINFALL IN PUNJAB',color='black',fontsize=23)
plt.xlabel('YEAR',color='black',fontsize=20)
plt.grid()
```

```python
data8=Punjab[['Jan-Feb','Mar-May','Jun-Sep','Oct-Dec','YEAR']].groupby('YEAR').sum()
data8.plot.line(figsize=(12,12),marker='o')
plt.title('QUARTERLY RAINFALL IN PUNJAB',color='black',fontsize=23)
plt.xlabel('YEAR',color='black',fontsize=20)
plt.grid()
```

```python
plot5=Punjab[['ANNUAL','YEAR']].groupby('YEAR').sum()
plot5.plot.line(figsize=(12,12),marker='o')
plt.xlabel('YEAR',color='red',fontsize=23)
plt.ylabel('ANNUAL',color='red',fontsize=23)
plt.title('ANNUAL RAINFALL IN PUNJAB',color='red',fontsize=25)
```

```python
# ANNUAL rainfall in india from 1915 to 2015
plt.figure(dpi=250)
# plt.title('ANNUAL RAINFALL')
data9=data[['ANNUAL','YEAR']].groupby('YEAR').sum()
data9.plot.line(marker='o',markerfacecolor='black',figsize=(20,12),color='red')
plt.title('ANNUAL RAINFALL',fontsize=24,color='black')
plt.xlabel('YEAR',color='black',fontsize=20)
plt.grid()
```
